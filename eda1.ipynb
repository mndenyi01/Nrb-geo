## Nairobi Sub-Counties Ploygon plot

- Dataset downloaded from `https://gadm.org/download_country.html`
- Extracting the zip file to get the '*.json' file and deposting  it in the data folder.
- Specifically the level 2 that is detailed till sub-counties

```python
import geopandas as gpd

gdf = gpd.read_file("./data/gadm41_KEN_2.json")
print(gdf.head())

gdf.shape
```

```python
# gdf.columns

nrb_df = gdf[gdf["NAME_1"] == "Nairobi"]

nrb_df.shape

print(nrb_df[["NAME_2"]])
```

```python
# Plotting the Map
nrb_df.plot(figsize=(10, 10), edgecolor='black')
```

```python
# Labelling the sub-counties
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 10))
nrb_df.plot(ax=ax, edgecolor="black")

for idx, row in nrb_df.iterrows():
    plt.text(row.geometry.centroid.x,
             row.geometry.centroid.y,
             row["NAME_2"],
             fontsize=8)

plt.title("Nairobi Sub-Counties")
plt.show()
```
